```python
#import libraries
import csv
import numpy as np
import matplotlib.pyplot as pyplot
```

```python
#read file
fin = 'Charlotte_FY_2021_BudgetReport.csv'
with open(fin, 'r') as budget:
    reader = csv.reader(budget, delimiter=',')
    data = np.array(list(reader))
budget.close()
```

```python
#remove extra columns
new_data = np.delete(data, [0,1,3,5,9],1)
print(new_data[0:5])
```

```python
#total CMPD Budget
CMPD_Total = 0
for row in new_data:
    if row[1] == '30':
        CMPD_Total = CMPD_Total + int(row[4])
print("CMPD's total expenses are: $" + str(CMPD_Total))
```

```python
#Subtract Departmental Charges listed in pdf (based on how other departments balance)
PubCMPD_Total_Exp = 290203220
CMPD_Adjusted_Total=CMPD_Total-618324
total_discrepancy = PubCMPD_Total_Exp-CMPD_Adjusted_Total
print("CMPD's published expenses are: $" + str(PubCMPD_Total_Exp))
print("CMPD's calculated adjusted expenses are: $" + str(CMPD_Adjusted_Total))
print("CMPD's total spending discrepancy is: $" + str(total_discrepancy))
```

```python
#Calculate Personnel Expenses
PubCMPD_Personnel_Exp = 247347250
CMPD_Personnel_Expenses=0
for row in new_data:
    if row[1] == '30':
        if row[2][:2] == '51':
            CMPD_Personnel_Expenses = CMPD_Personnel_Expenses + int(row[4])
personnel_discrepancy = PubCMPD_Personnel_Exp - CMPD_Personnel_Expenses
print("CMPD's published personnel expenses are: $" + str(PubCMPD_Personnel_Exp))
print("CMPD's calculated personnel expenses are: $" + str(CMPD_Personnel_Expenses))
print("CMPD's personnel spending discrepancy is: $" + str(personnel_discrepancy))
```

```python
#Calculate Operating Expenses
PubCMPD_Operating_Exp = 43464294
CMPD_Operating_Expenses=0
for row in new_data:
    if row[1] == '30':
        if row[2][:2] == '52':
            CMPD_Operating_Expenses = CMPD_Operating_Expenses + int(row[4])
operating_discrepancy = PubCMPD_Operating_Exp - CMPD_Operating_Expenses
print("CMPD's published operating expenses are: $" + str(PubCMPD_Operating_Exp))
print("CMPD's calculated operating expenses are: $" + str(CMPD_Operating_Expenses))
print("CMPD's operating spending discrepancy is: $" + str(operating_discrepancy))
```

```python
#Calculate Capital Outlay
PubCMPD_Capital_Outlay = 10000
CMPD_Capital_Outlay=0
for row in new_data:
    if row[1] == '30':
        if row[2][:2] == '53':
            CMPD_Capital_Outlay = CMPD_Capital_Outlay + int(row[4])
capital_outlay_discrepancy = PubCMPD_Capital_Outlay - CMPD_Capital_Outlay
print("CMPD's published capital outlay expenses are: $" + str(PubCMPD_Capital_Outlay))
print("CMPD's calculated capital outlay expenses are: $" + str(CMPD_Capital_Outlay))
print("CMPD's capital outlay discrepancy is $" + str(capital_outlay_discrepancy))
```

```python
#Compare to Fire
#total Fire Budget
Fire_Total = 0
for row in new_data:
    if row[1] == '31':
        Fire_Total = Fire_Total + int(row[4])
print("Fire's total expenses are: $" + str(Fire_Total))

#Subtract Departmental Charges listed in pdf (based on how other departments balance)
PubFire_Total_Exp = 137741974
Fire_Adjusted_Total=Fire_Total-284793
fire_total_discrepancy = PubFire_Total_Exp-Fire_Adjusted_Total
print("Fire's published expenses are: $" + str(PubFire_Total_Exp))
print("Fire's calculated adjusted expenses are: $" + str(Fire_Adjusted_Total))
print("Fire's total spending discrepancy is: $" + str(fire_total_discrepancy))

#Calculate Personnel Expenses
PubFire_Personnel_Exp = 122320352
Fire_Personnel_Expenses=0
for row in new_data:
    if row[1] == '31':
        if row[2][:2] == '51':
            Fire_Personnel_Expenses = Fire_Personnel_Expenses + int(row[4])
fire_personnel_discrepancy = PubFire_Personnel_Exp - Fire_Personnel_Expenses
print("Fire's published personnel expenses are: $" + str(PubFire_Personnel_Exp))
print("Fire's calculated personnel expenses are: $" + str(Fire_Personnel_Expenses))
print("Fire's personnel spending discrepancy is: $" + str(fire_personnel_discrepancy))

#Calculate Operating Expenses
PubFire_Operating_Exp = 15706415
Fire_Operating_Expenses=0
for row in new_data:
    if row[1] == '31':
        if row[2][:2] == '52' or row[2][:2] == '57':
            Fire_Operating_Expenses = Fire_Operating_Expenses + int(row[4])
fire_operating_discrepancy = PubFire_Operating_Exp - Fire_Operating_Expenses
print("Fire's published operating expenses are: $" + str(PubFire_Operating_Exp))
print("Fire's calculated operating expenses are: $" + str(Fire_Operating_Expenses))
print("Fire's operating spending discrepancy is: $" + str(fire_operating_discrepancy))
```

```python
#Filter just CMPD data
CMPD_data = new_data[new_data [:,1] == '30']
print(CMPD_data)
```

```python
#get categories
unique_list = []
for row in CMPD_data:
    if row[3] not in unique_list:
        unique_list.append(row[3])
print(unique_list)
```

```python
#remove codes from categories
category_list=[]
code_list=[]
for item in unique_list:
    category_list.append(item[7:])
    code_list.append(item[:6])
print(code_list)
```

```python
#sum by category
category_sums = []
for code in code_list:
    cat_sum = 0
    cat_data = CMPD_data[CMPD_data [:,2] == code]
    for row in cat_data:
        cat_sum += int(row[4])
    category_sums.append(cat_sum)
print(category_sums)

```

```python
#separate categories
personnel_categories = category_list[:14]
personnel_costs = category_sums[:14]
operational_categories = category_list[14:-1]
operational_costs = category_sums[14:-1]
capital_outlay_category = category_list[-1]
capital_outlay_costs = category_sums[-1]
print(operational_categories)
```

```python
#Create lists for plotting
budget_sections = ['Total', 'Personnel', 'Operating', 'Capital']
CMPD_published_list = [PubCMPD_Total_Exp, PubCMPD_Personnel_Exp, PubCMPD_Operating_Exp, PubCMPD_Capital_Outlay]
CMPD_calculated_list = [CMPD_Adjusted_Total, CMPD_Personnel_Expenses, CMPD_Operating_Expenses, CMPD_Capital_Outlay]
```

```python
#Graph Budget Discrepancy
pyplot.figure(figsize = (10, 10))
indices = np.arange(len(budget_sections))
width = 0.20
bar1=pyplot.bar(indices, CMPD_published_list, width=width, label='Published')
bar2=pyplot.bar(indices+width, CMPD_calculated_list, width=width, label='Calculated')
pyplot.xticks(ticks=indices, labels=budget_sections)
pyplot.xlabel('Budget Categories', fontsize=20)
pyplot.ylabel('Amount (in hundred thousands)', fontsize=20)
pyplot.title('CMPD Budget Published vs Calculated', fontsize=26)
pyplot.legend()
def addlabels1(x,y):
    for i in range(len(x)):
        pyplot.text(i,y[i],y[i],horizontalalignment='left', verticalalignment='bottom', color='orange')
def addlabels2(x,y):
    for i in range(len(x)):
        pyplot.text(i,y[i],y[i],horizontalalignment='right', verticalalignment='bottom', color='blue')
addlabels1(budget_sections, CMPD_calculated_list)
addlabels2(budget_sections, CMPD_published_list)
pyplot.show()
```

```python
#Graph Budget Breakdown
pyplot.figure(figsize = (10, 10))
pyplot.pie(CMPD_calculated_list[1:],labels = budget_sections[1:],textprops={'fontsize': 18},autopct='%1.2f%%')
pyplot.title('CMPD Total Budget Distribution', fontsize=24)
pyplot.show()

pyplot.figure(figsize = (10, 10))
pyplot.barh(personnel_categories,personnel_costs, color='orange')
pyplot.title("CMPD Personnel Budget Categories", fontsize=30)
pyplot.ylabel("Category", fontsize=16)
pyplot.xlabel("Amount (in hundred thousands)", fontsize=16)
pyplot.show()

pyplot.figure(figsize = (10, 10))
pyplot.barh(operational_categories,operational_costs)
pyplot.title("CMPD Operational Budget Categories", fontsize=30)
pyplot.ylabel("Category", fontsize=16)
pyplot.xlabel("Amount", fontsize=16)
pyplot.show()
```
